```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        # Create a new Point by adding the x and y of both points
        return Point(self.x + other.x, self.y + other.y)

    def __sub__(self,other):
        return Point(self.x-other.x,self.y-other.y)

    def __repr__(self):
        return f"Point({self.x}, {self.y})"
    
    def __str__(self):
        pass

# Example usage
p1 = Point(2, 3)
p2 = Point(4, 5)

result = p1 + p2
print(result)  # Output: Point(6, 8)
```

```python
class Person:
  def __init__(self, name, age=18): # constructor
    self.name = name
    self.age = age

  def __str__(self):
    return f"{self.name}"
  
  def __repr__(self):
    return f"{self.name} {self.age} object"
  


p1 = Person("John", 36)

p2=Person("Gokul")

print(p1)
print(p2)
```
