```python
def binary_search(l,target):
    n=len(l)
    left,right=0,n-1
    count=0
    while left<=right:
        mid=(left+right)//2
        count+=1
        if l[mid]==target:
            return (True,count)
        elif l[mid]<target:
            left=mid+1
        else:
            right=mid-1
            
    return (False,count)

l=[2, 6, 8, 11, 17, 23, 33, 44, 46, 50, 65]
binary_search(l,100)
```
